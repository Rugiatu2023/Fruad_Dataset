```python
import os
import requests
import streamlit as st

API_URL = os.getenv("API_URL", "http://localhost:8000")

st.title("Fraud Detection Demo")
st.caption(f"API: {API_URL}")

# Match the same features as in API
FEATURES = ["feature1", "feature2", "feature3"]

st.subheader("Input features")
vals = []
for f in FEATURES:
    vals.append(st.number_input(f, value=0.0))

if st.button("Predict"):
    try:
        r = requests.post(
            f"{API_URL}/predict",
            json={"values": vals},
            timeout=15,
        )
        r.raise_for_status()
        out = r.json()

        st.success(f"Prediction: {out['prediction']} (1 = fraud, 0 = not fraud)")
        if out.get("proba_fraud") is not None:
            st.write(f"Fraud probability: {out['proba_fraud']:.4f}")

    except Exception as e:
        st.error(f"Request failed: {e}")
```
